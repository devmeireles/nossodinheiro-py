```python
# Functions import

from helpers.download_file import download_and_extract_file
from helpers.format_to_millions import format_to_millions
from helpers.get_sum_data import get_sum_data

import os
import pandas as pd
```

```python
# Download the dataset

url = "https://www.camara.leg.br/cotas/Ano-2023.csv.zip"
download_and_extract_file(url, output_dir="data")
```

```python
df = pd.read_csv('./data/Ano-2023.csv', delimiter=";")
df.head()
```

```python
txtDescricaoCusto = get_sum_data(df, ['txtDescricao'], 'vlrDocumento', 'Custo', 'Custo')
txtDescricaoCusto.head(10)
```

```python
sgPartidoCusto = get_sum_data(df, ['sgPartido'], 'vlrDocumento', 'Custo', 'Custo')
sgPartidoCusto.head(10)
```

```python
txNomeParlamentarCusto = get_sum_data(df, ['txNomeParlamentar', 'sgPartido'], 'vlrDocumento', 'Custo', 'Custo')
txNomeParlamentarCusto.head(10)
```

```python
supplier_mapping = {
  'Cia Aérea - GOL': 'GOL',
  'Cia Aérea - TAM': 'TAM'
}

df['txtFornecedor'] = df['txtFornecedor'].replace(supplier_mapping)
txtFornecedorCusto = get_sum_data(df, ['txtFornecedor'], 'vlrDocumento', 'Custo', 'Custo')
txtFornecedorCusto.head(10)
```

```python
grouped_data = df.groupby('numMes')['vlrDocumento'].sum()
top_party_df = grouped_data.reset_index(name='TotalExpense').sort_values(by='TotalExpense', ascending=False)

top_party_df['TotalExpense'] = top_party_df['TotalExpense'].apply(format_to_millions)

top_party_df.head(10)
```
